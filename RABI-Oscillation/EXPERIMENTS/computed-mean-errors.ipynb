```python
the = [THE_H1, THE_H1, THE_H1, THE_H2, THE_H3]
exp = [NOI_H1, MIT_H1, IBM_H1, IBM_H2, IBM_H3]

i = 4

THE = the[i]
EXP = exp[i]

SUM = 0

for i in range(1,50):
    if(THE[0].loc["run-1"][i] < 0.1):
        SUM += abs(EXP[0].loc["mean"][i] - THE[0].loc["run-1"][i]) / abs(THE[0].loc["run-1"][i] - 1)
    else:
        SUM += abs(EXP[0].loc["mean"][i] - THE[0].loc["run-1"][i]) / abs(THE[0].loc["run-1"][i])


print("H1 in IBMQ:", SUM / 50 * 100, "%")
```
